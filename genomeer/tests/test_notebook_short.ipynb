```python
from genomeer.agent import BioAgent

agent = BioAgent(path="./data",
    llm="gpt-oss:20b",
    source="Custom",
    use_tool_retriever=True,
    timeout_seconds=600,
    base_url="http://10.52.88.30:11434/v1",
    api_key=None
)
```

```python
response = agent.go('Hi BioAgent, can you give me a list of all tolls, library and software and task you can help me with?')
```

```python
response = agent.go('What can you help me with in terms of visualization tasks?')
```
