```python
import os, time
from pathlib import Path
from genomeer.utils.stream.shared import REGISTRY 
```

```python
def listen(stream_id: str):
    # Get the persisted file path (works even if this process never created the stream)
    path: Path = REGISTRY.attach_reader(stream_id)
    # wait for producer to create the file
    while not path.exists():
        time.sleep(0.1)

    with open(path, "r", encoding="utf-8", errors="replace") as f:
        while True:
            line = f.readline()
            if not line:
                time.sleep(0.1)
                continue
            line = line.rstrip("\n")
            if line.startswith("[EPH] "):
                # ephemeral snapshot: show as carriage-return update if you like
                print("\r" + line[6:], end="", flush=True)
                continue
            print(line)
            if line == "<<CLOSED>>":
                print("\n<< stream closed >>")
                break
```

```python
listen("4c550784213445ba93302d3030f29b38")
```
