```python
from genomeer.agent import BioAgent
```

```python
agent = BioAgent(path="./data",
    llm="gpt-oss:20b",
    source="Custom",
    use_tool_retriever=True,
    timeout_seconds=600,
    base_url="http://10.52.88.30:11434/v1",
    api_key=None
)
```

```python
response = agent.go('I wanna do sequence alignment, can you help me?')
```

```python
response = agent.go('can you execute this code python: caculate fibnacci of 2000')
```

```python
from IPython.display import display, clear_output

lines = []
for step in agent.go_stream("can you execute this code python: calculate fibonacci of 2000"):
    lines.append(step["output"])
    clear_output(wait=True)
    display("\n".join(lines))
```
